## **parte 1)** **Crear una base de datos en SQLite utilizando la libreria SQLalchemy. https://docs.sqlalchemy.org/en/14/core/engines.html#sqlite**


Para crear una base de datos en SQLite utilizando SQLAlchemy, primero hay que importar la librería y luego utilizar la función create_engine para establecer una conexión con la base de datos.

```python
pip install sqlalchemy
```

```python
import pandas as pd
from sqlalchemy import create_engine, MetaData, Table, Column, Integer, String, Float

# Paso 1: Leer los datos desde el CSV
melb_df = pd.read_csv('https://cs.famaf.unc.edu.ar/~mteruel/datasets/diplodatos/melb_data.csv')

# Paso 2: Crear una conexión a la base de datos SQLite
engine = create_engine('sqlite:///melbourne_housing.sqlite')

# Paso 3: Crear una tabla en la base de datos
metadata = MetaData()

housing_table = Table('housing', metadata,
    Column('index', Integer, primary_key=True),
    Column('Suburb', String),
    Column('Address', String),
    Column('Rooms', Integer),
    Column('Type', String),
    Column('Price', Float),
    Column('Method', String),
    Column('SellerG', String),
    Column('Date', String),
    Column('Distance', Float),
    Column('Postcode', Integer),
    Column('Bedroom2', Float),
    Column('Bathroom', Float),
    Column('Car', Float),
    Column('Landsize', Float),
    Column('BuildingArea', Float),
    Column('YearBuilt', Float),
    Column('CouncilArea', String),
    Column('Lattitude', Float),
    Column('Longtitude', Float),
    Column('Regionname', String),
    Column('Propertycount', Float)
)

metadata.create_all(engine)  # Crea la tabla en la base de datos

# Paso 4: Insertar los datos en la tabla
with engine.connect() as connection:
    melb_df.to_sql('housing', con=connection, if_exists='replace', index_label='index')

print("Datos insertados en la base de datos SQLite.")
```

a) Importar las librerías necesarias:
- pandas para manejar los datos en formato DataFrame.
- create_engine, MetaData, Table, Column, Integer, String, Float de SQLAlchemy para crear y gestionar la base de datos SQLite.

b) pandas.read_csv() lee los datos desde la URL proporcionada y generar un DataFrame.

c)  create_engine('sqlite:///nombre_de_la_base_de_datos.sqlite') crea una conexión a la base de datos SQLite.

d) metadata.create_all(engine) crea la tabla en la base de datos,  de acuerdo con las columnas del DataFrame

e) to_sql de pandas es para insertar los datos del DataFrame en la tabla. El argumento if_exists='replace' asegura que si la tabla ya existe, se reemplazará.

# **Parte 2) Ingestar los datos provistos en 'https://cs.famaf.unc.edu.ar/~mteruel/datasets/diplodatos/melb_data.csv' en una tabla y el dataset generado en clase con datos de airbnb y sus precios por codigo postal en otra.**

```python
import pandas as pd
from sqlalchemy import create_engine, MetaData, Table, Column, Integer, String, Float

# Paso 1: Leer los datos desde el CSV de Melbourne housing
melb_df = pd.read_csv('https://cs.famaf.unc.edu.ar/~mteruel/datasets/diplodatos/melb_data.csv')

# Paso 1b: Leer los datos desde el CSV de Airbnb
interesting_cols = [
    'description', 'neighborhood_overview',
    'street', 'neighborhood', 'city', 'suburb', 'state', 'zipcode',
    'price', 'weekly_price', 'monthly_price',
    'latitude', 'longitude',
]
airbnb_df = pd.read_csv(
    'https://cs.famaf.unc.edu.ar/~mteruel/datasets/diplodatos/cleansed_listings_dec18.csv',
    usecols=interesting_cols
)

# Paso 2: Crear una conexión a la base de datos SQLite
engine = create_engine('sqlite:///melbourne_housing.sqlite')

# Paso 3: Crear las tablas en la base de datos
metadata = MetaData()

# Tabla para los datos de Melbourne housing
housing_table = Table('housing', metadata,
    Column('index', Integer, primary_key=True),
    Column('Suburb', String),
    Column('Address', String),
    Column('Rooms', Integer),
    Column('Type', String),
    Column('Price', Float),
    Column('Method', String),
    Column('SellerG', String),
    Column('Date', String),
    Column('Distance', Float),
    Column('Postcode', Integer),
    Column('Bedroom2', Float),
    Column('Bathroom', Float),
    Column('Car', Float),
    Column('Landsize', Float),
    Column('BuildingArea', Float),
    Column('YearBuilt', Float),
    Column('CouncilArea', String),
    Column('Lattitude', Float),
    Column('Longtitude', Float),
    Column('Regionname', String),
    Column('Propertycount', Float)
)

# Tabla para los datos de Airbnb
airbnb_table = Table('airbnb', metadata,
    Column('index', Integer, primary_key=True),
    Column('description', String),
    Column('neighborhood_overview', String),
    Column('street', String),
    Column('neighborhood', String),
    Column('city', String),
    Column('suburb', String),
    Column('state', String),
    Column('zipcode', String),
    Column('price', Float),
    Column('weekly_price', String),
    Column('monthly_price', String),
    Column('latitude', Float),
    Column('longitude', Float)
)

metadata.create_all(engine)  # Crea las tablas en la base de datos

# Paso 4: Insertar los datos en las tablas
with engine.connect() as connection:
    melb_df.to_sql('housing', con=connection, if_exists='replace', index_label='index')
    airbnb_df.to_sql('airbnb', con=connection, if_exists='replace', index_label='index')

print("Datos insertados en la base de datos SQLite.")
```

Se insertan los datos del DataFrame de Airbnb en la tabla airbnb.

Con estos pasos, se crean dos tablas en la base de datos SQLite y cargado datos tanto de Melbourne housing como de Airbnb. Este enfoque asegura que los datos de ambas fuentes se almacenen correctamente en la base de datos.

En el siguiente código verificamos la correcta creacion de ambas base de datos haciendo simples consultas como el tamaño de las tablas y visualizar los primeros 5 registros

```python
import pandas as pd
from sqlalchemy import create_engine

# Crear una conexión a la base de datos SQLite
engine = create_engine('sqlite:///melbourne_housing.sqlite')

# Consultas SQL
query_size_housing = "SELECT COUNT(*) FROM housing;"
query_size_airbnb = "SELECT COUNT(*) FROM airbnb;"
query_sample_housing = "SELECT * FROM housing LIMIT 5;"
query_sample_airbnb = "SELECT * FROM airbnb LIMIT 5;"

# Ejecutar las consultas y obtener los resultados
with engine.connect() as connection:
    # Obtener el tamaño de la tabla housing
    size_housing = pd.read_sql(query_size_housing, connection).iloc[0, 0]
    # Obtener el tamaño de la tabla airbnb
    size_airbnb = pd.read_sql(query_size_airbnb, connection).iloc[0, 0]

    # Obtener las primeras 5 filas de la tabla housing
    sample_housing = pd.read_sql(query_sample_housing, connection)

    # Obtener las primeras 5 filas de la tabla airbnb
    sample_airbnb = pd.read_sql(query_sample_airbnb, connection)

# Mostrar los resultados
print(f"Tamaño de la tabla 'housing': {size_housing} registros")
print(f"Tamaño de la tabla 'airbnb': {size_airbnb} registros")

print("\nPrimeras 5 filas de la tabla 'housing':")
print(sample_housing)

print("\nPrimeras 5 filas de la tabla 'airbnb':")
print(sample_airbnb)
```

# **Parte 3) Implementar consultas en SQL que respondan con la siguiente información:**

- cantidad de registros totales por ciudad.
- cantidad de registros totales por barrio y ciudad.

```python
import pandas as pd
from sqlalchemy import create_engine

# Crear una conexión a la base de datos SQLite
engine = create_engine('sqlite:///melbourne_housing.sqlite')

# Consultas SQL para obtener la cantidad de registros totales por ciudad y por barrio y ciudad
# Para la tabla housing
query_total_por_ciudad_housing = """
SELECT CouncilArea AS city, COUNT(*) AS total_registros
FROM housing
GROUP BY CouncilArea;
"""

query_total_por_barrio_ciudad_housing = """
SELECT CouncilArea AS city, Suburb AS neighborhood, COUNT(*) AS total_registros
FROM housing
GROUP BY CouncilArea, Suburb;
"""

# Para la tabla airbnb
query_total_por_ciudad_airbnb = """
SELECT city, COUNT(*) AS total_registros
FROM airbnb
GROUP BY city;
"""

query_total_por_barrio_ciudad_airbnb = """
SELECT city, neighborhood, COUNT(*) AS total_registros
FROM airbnb
GROUP BY city, neighborhood;
"""

# Ejecutar las consultas y obtener los resultados
with engine.connect() as connection:
    # Ejecutar las consultas para la tabla housing
    df_total_por_ciudad_housing = pd.read_sql(query_total_por_ciudad_housing, connection)
    df_total_por_barrio_ciudad_housing = pd.read_sql(query_total_por_barrio_ciudad_housing, connection)

    # Ejecutar las consultas para la tabla airbnb
    df_total_por_ciudad_airbnb = pd.read_sql(query_total_por_ciudad_airbnb, connection)
    df_total_por_barrio_ciudad_airbnb = pd.read_sql(query_total_por_barrio_ciudad_airbnb, connection)

# Mostrar los resultados
print("Cantidad de registros totales por ciudad en la tabla 'housing':")
print(df_total_por_ciudad_housing)

print("\nCantidad de registros totales por barrio y ciudad en la tabla 'housing':")
print(df_total_por_barrio_ciudad_housing)

print("\nCantidad de registros totales por ciudad en la tabla 'airbnb':")
print(df_total_por_ciudad_airbnb)

print("\nCantidad de registros totales por barrio y ciudad en la tabla 'airbnb':")
print(df_total_por_barrio_ciudad_airbnb)
```

# **Parte 4) Combinar los datasets de ambas tablas ingestadas utilizando el comando JOIN de SQL para obtener un resultado similar a lo realizado con Pandas en clase.**

```python
import pandas as pd
from sqlalchemy import create_engine

# Crear una conexión a la base de datos SQLite
engine = create_engine('sqlite:///melbourne_housing.sqlite')

# Leer el DataFrame de airbnb
airbnb_df = pd.read_csv(
    'https://cs.famaf.unc.edu.ar/~mteruel/datasets/diplodatos/cleansed_listings_dec18.csv',
    usecols=['zipcode']
)

# Modificar el tipo de datos de la columna zipcode
airbnb_df['zipcode'] = pd.to_numeric(airbnb_df['zipcode'], errors='coerce')

# Eliminar NaN en la columna zipcode y convertir a entero
airbnb_df['zipcode_int'] = airbnb_df['zipcode'].fillna(0).astype(int)

# Leer el DataFrame de viviendas de la base de datos SQLite
housing_df = pd.read_sql_table('housing', con=engine)

# Encontrar la intersección de zipcodes entre airbnb_df y housing_df
intersection = pd.Series(list(set(airbnb_df['zipcode_int']).intersection(set(housing_df['Postcode']))))

print("\nIntersección de zipcodes entre airbnb_df y housing_df:")
print(intersection)

# Imprimir información adicional sobre los zipcodes
print("\nAirbnb zipcodes únicos:", len(airbnb_df['zipcode_int'].unique()))
print("Melbourne Housing zipcodes únicos:", len(housing_df['Postcode'].unique()))
print("Zipcodes comunes:", len(intersection))
```

```python
# Consulta SQL para combinar los datasets utilizando INNER JOIN
query = """
SELECT *
FROM airbnb
INNER JOIN housing ON airbnb.zipcode = housing.Postcode
"""

# Ejecutar la consulta y cargar los resultados en un DataFrame
combined_df = pd.read_sql_query(query, engine)

# Imprimir las primeras filas del DataFrame resultante
print(combined_df.head())
```
